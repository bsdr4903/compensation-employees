```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("##raw_data.csv", index_col=[0])
```

```python
temp = df[ df['Industry'].isin(
    ['TOTAL', ' Agriculture, hunting, forestry and fishing [A]',
       ' Mining and quarrying [B]', ' Manufacturing [C]',
       ' Electricity, gas, steam and air conditioning supply [D]',
       ' Water supply; sewerage, waste management and remediation activities [E]',
       ' Construction [F]',
       ' Wholesale and retail trade, repair of motor vehicles and motorcycles [G]',
       ' Transportation and storage [H]',
       ' Accommodation and food service activities [I]',
       ' Information and communication [J]',
       ' Financial and insurance activities [K]',
       ' Real estate activities [L]',
       ' Professional, scientific and technical activities [M]',
       ' Administrative and support service activities [N]',
       ' Public administration and defence; compulsory social security [O]',
       ' Education [P]', ' Human health and social work activities [Q]',
       ' Arts, entertainment and recreation [R]',
       ' Other service activities [S]',
       ' Activities of households as employers; undifferentiated activities of households for own use [T]']
    ) ]
temp
```

```python
# Create a dictionary mapping industries to sectors
sector_mapping = {
    ' Agriculture, hunting, forestry and fishing [A]': 'Agriculture',
    ' Mining and quarrying [B]': 'Manufacturing',
    ' Manufacturing [C]': 'Manufacturing',
    ' Construction [F]': 'Manufacturing',
    ' Electricity, gas, steam and air conditioning supply [D]': 'Service',
    ' Water supply; sewerage, waste management and remediation activities [E]': 'Service',
    ' Wholesale and retail trade, repair of motor vehicles and motorcycles [G]': 'Service',
    ' Transportation and storage [H]': 'Service',
    ' Accommodation and food service activities [I]': 'Service',
    ' Information and communication [J]': 'Service',
    ' Financial and insurance activities [K]': 'Service',
    ' Real estate activities [L]': 'Service',
    ' Professional, scientific and technical activities [M]': 'Service',
    ' Administrative and support service activities [N]': 'Service',
    ' Public administration and defence; compulsory social security [O]': 'Service',
    ' Education [P]': 'Service', ' Human health and social work activities [Q]': 'Service',
    ' Arts, entertainment and recreation [R]': 'Service',
    ' Other service activities [S]': 'Service',
    ' Activities of households as employers; undifferentiated activities of households for own use [T]': 'Service'
}

# Map industries to sectors
temp['Sector'] = temp['Industry'].map(sector_mapping)
temp

# Group by Country, Time, and Sector, then sum the 'Share' values
sector_share = temp.groupby(['Country', 'Time', 'lnGDP', 'Sector'])['Labour Costs Share(%)'].sum().reset_index()

# Display the aggregated 'Share' data for each sector
sector_share
```

```python
sector_share = sector_share.set_index(['Country', 'Time'])
sector_share
```
